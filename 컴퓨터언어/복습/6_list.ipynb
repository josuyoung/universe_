```python
L = ['7:27','14:22','6:9','6:11','2:19']
```

```python
def mykey(lel):
    return [int(s) for s in lel.split(":")]

#     return int(lel.split(':')[0]),int(lel.split(':')[1])
L.sort(key=mykey)
L
```

```python
def mykey(lel):
    return sum([int(s) for s in lel.split(":")])
#합친값으로 비교
L.sort(key=mykey)
L
```

lambda function(람다 함수)

```python
f = lambda e : e*e
f(3)
```

```python
L.sort(key = lambda ele: sum([int(s) for s in ele.split(":")]) )
```
